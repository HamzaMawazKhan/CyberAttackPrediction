# Determining the top 5 correlated features


```python
# Flatten the correlation matrix and reset index to convert the matrix to a three-column DataFrame
corr_pairs = correlation_matrix.abs().unstack().reset_index()

# Name the columns for clarity
corr_pairs.columns = ['Feature1', 'Feature2', 'Correlation']

# Remove self-correlations (where Feature1 is the same as Feature2)
corr_pairs = corr_pairs[corr_pairs['Feature1'] != corr_pairs['Feature2']]

# Sort by correlation, descending
corr_pairs_sorted = corr_pairs.sort_values(by='Correlation', ascending=False)

# Get the top 5 off-diagonal correlation pairs
top_5_corr_pairs = corr_pairs_sorted.head(10)  # Get top 10 because pairs are mirrored across the diagonal

# Since the correlation matrix is symmetric, the pairs are duplicated (A, B) and (B, A), so we take every other pair
top_5_unique_corr_pairs = top_5_corr_pairs.iloc[::2, :]

print(top_5_unique_corr_pairs)
```

# Neural Network Model Description:
- This is a Sequential model, a linear stack of layers.
- Input Layer: Dense layer with 20 neurons, 'relu' activation, expecting 28 input features.
- Intermediate Layers:
   - LeakyReLU layer with alpha=0.05 following the first Dense layer.
   - Second Dense layer with 30 neurons, 'relu' activation, followed by another LeakyReLU layer with alpha=0.05.
   - Third Dense layer with 40 neurons, 'relu' activation.
- Output Layer: Dense layer with 5 neurons, 'softmax' activation for multiclass classification.
- Model compiled with Adam optimizer (lr=0.01), using 'categorical_crossentropy' loss function and tracking 'accuracy' metric.
- This architecture is designed for a multiclass classification problem, with layers building up complexity and a softmax output for probability distribution across classes.


```python
modelAnn = Sequential()
modelAnn.add(Dense(20, input_dim=28, activation='relu'))
modelAnn.add(LeakyReLU(alpha=0.05))
modelAnn.add(Dense(30, activation='relu'))
modelAnn.add(LeakyReLU(alpha=0.05))
modelAnn.add(Dense(40, activation='relu'))
modelAnn.add(Dense(5,activation='softmax'))

modelAnn.compile(Adam(lr=0.01), loss = "categorical_crossentropy", metrics = ['accuracy'])

print(modelAnn.summary())
```

# Training and Callback Configuration:
- `ModelCheckpoint` is set to save the model as 'best_model3.hdf5'. It monitors the 'loss' metric, saves only the best model (in terms of loss), and checks after each epoch.
- `EarlyStopping` monitors 'val_loss' (validation loss). If the validation loss doesn't improve for 3 consecutive epochs ('patience' = 3), training will stop early to prevent overfitting.
- `modelAnn.fit` is used to train the model. 
   - Data: `train_x` (features) and `train_y` (labels).
   - Validation Split: 20% of the data is used for validation.
   - Epochs: Maximum number of epochs set to 50. Training may stop early due to EarlyStopping.
   - Batch Size: 32, determines the number of samples processed before the model is updated.
   - Verbose: 1, to show a progress bar during training.
   - Callbacks: The defined `ModelCheckpoint` and `EarlyStopping` are used to monitor and enhance training.
- This setup aims to optimize the model by balancing between sufficient training and avoiding overfitting, using validation loss as a guide.


```python
checkpoint = ModelCheckpoint("best_model3.hdf5", monitor='loss', verbose=1, save_best_only=True, mode='auto', period=1)
earlystopping = EarlyStopping(monitor = 'val_loss', patience = 3, mode = 'auto')

history = modelAnn.fit(train_x, train_y, validation_split= 0.20, epochs = 50, verbose = 1, batch_size= 32, callbacks=[checkpoint, earlystopping])
```

```python
ANN_Model_score = modelAnn.evaluate(test_x, test_y, verbose=0)
print('Test loss:', ANN_Model_score[0])
print('Test accuracy:', ANN_Model_score[1])
Accuracy_ANN = ANN_Model_score[1]*100
```

# RandomForestClassifier Training and Prediction:
- `clf` is an instance of RandomForestClassifier with 100 trees (`n_estimators=100`).
- The model is trained using the `fit` method on the training data `train_x` (features) and `train_y` (labels).
- After training, the model is used to make predictions on the test dataset `test_x` using the `predict` method.
- These predictions are stored in `pred_RFC`.
- The accuracy of the model is then calculated by comparing the predicted values `pred_RFC` with the actual test labels `test_y`.
- `metrics.accuracy_score` is used to compute the accuracy percentage, which is then multiplied by 100 for a percentage format.
- The final accuracy is printed out as "Accuracy: [value]%".
- This approach provides a straightforward evaluation of the RandomForestClassifier's performance on the test data.


```python
clf=RandomForestClassifier(n_estimators=100)
clf.fit(train_x,train_y)
pred_RFC = clf.predict(test_x)

Accuracy_RandomFC = metrics.accuracy_score(test_y, pred_RFC)*100
print("Accuracy:", Accuracy_RandomFC)
```

# Logistic Regression for Multi-Class Classification:
- The one-hot encoded labels for training and test datasets are converted back to single labels using `np.argmax`.
  `train_y_single_label` and `test_y_single_label` represent these single label versions for training and test datasets respectively.
- A Logistic Regression model (`logreg`) is created, specifically configured for multi-class classification using the 'multinomial' option.
- The logistic regression model is trained using the `fit` method on `train_x` (features) and `train_y_single_label` (single-label class targets).
- The model then makes predictions on the test dataset `test_x` which are stored in `logreg_predictions`.
- The accuracy of the model is evaluated by comparing `logreg_predictions` with `test_y_single_label` using the `accuracy_score` function.
- Additionally, `classification_report` is used to provide a detailed analysis of the model's performance, including precision, recall, and F1-score for each class.
- The accuracy is then printed out, and `Accuracy_LogReg` stores the accuracy percentage value.
- This setup is ideal for scenarios where multi-class classification is needed and interpretable models like logistic regression are preferred.


```python
# Convert one-hot encoded labels to single labels for multi-class classification
train_y_single_label = np.argmax(train_y, axis=1)
test_y_single_label = np.argmax(test_y, axis=1)

# Create a Logistic Regression model
logreg = LogisticRegression(multi_class='multinomial', solver='lbfgs')

# Train the model
logreg.fit(train_x, train_y_single_label)

# Make predictions
logreg_predictions = logreg.predict(test_x)

# Evaluate the model
print("Accuracy:", accuracy_score(test_y_single_label, logreg_predictions))
print(classification_report(test_y_single_label, logreg_predictions))

Accuracy_LogReg = accuracy_score(test_y_single_label, logreg_predictions)*100
```
